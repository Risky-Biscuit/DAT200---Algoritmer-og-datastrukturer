```python

class Node:
	def __init__(self, key):
		self.key = key        # the value stored in this node
		self.left = None      # left child (smaller values)
		self.right = None     # right child (larger values)

# A utility function to do inorder traversal of BST
def inorder(root):
	if root is not None:
		inorder(root.left)
		print(root.key, end=' ')
		inorder(root.right)

# A utility function to insert a new node with given key in BST
def insert(node, key):
	# If the tree is empty, return a new node
	if node is None:
		return Node(key)
	# Otherwise, recur down the tree
	if key < node.key:
		node.left = insert(node.left, key)    # go left for smaller values
	else:
		node.right = insert(node.right, key)  # go right for larger values
  # if the key already exists, do nothing

	# return the (unchanged) node pointer
	return node


def count_leaf_nodes(root):
    """ Let us take the following BST
      50
    / \
    30	 70
    / \ / \
  20 40 60 80

  Then it is expected to print 4 by calling the function
  """
		# write your code here
    # if the tree is empty, there are 0 leaf nodes
    if root is None:
      return 0
    
    # if this node has no children, it's a leaf -> count 1
    if root.left is None and root.right is None:
      return 1
    
    # otherwise, count leaves in left subtree + right subtree
    return count_leaf_nodes(root.left) + count_leaf_nodes(root.right)

  # end

def preorder(root):
    """ Let us take following BST
      50
    / \
    30	 70
    / \ / \
  20 40 60 80

  Then it is expected to print 50,30,20,40,70,60,80 by calling the function
  """
	# write your code here
    if root is None:
      return
    
    # 1. visit root
    print(root.key, end=" ")
    
    # 2. Go left
    preorder(root.left)
    
    # 3. Go right
    preorder(root.right)

  # end

def print_nodes_with_two_child(root):

    """ Let us take following BST
      50
    / \
    30	 70
    / \ / \
  20 40 60 80

  Then it is expected to print 50,30,70 by calling the function
  """
	# write your code here
    if root is None:
      return
    
    # Check this node
    if root.left is not None and root.right is not None:
      print(root.key, end=" ")
      
    # Recurse on children
    print_nodes_with_two_child(root.left)
    print_nodes_with_two_child(root.right)

  # end



def print_nodes_bylevel(root):

    """ Let us take following BST
      50
    / \
    30	 70
    / \ / \
  20 40 60 80

  Then it is expected to print 50,30,70,20,40,60,80 by calling the function
  """
  # write your code here
    if root is None:
      return
    
    queue = [root]  # start with the root in a list
    
    while queue:
      current = queue.pop(0)        # take first element
      print(current.key, end=" ")   # process it
      
      # add its children to the queue (left first, then right)
      if current.left is not None:
        queue.append(current.left)
      if current.right is not None:
        queue.append(current.right)

  # end




def tree_height(root):
    # write your code here
    if root is None:
      return -1   # empty tree has height -1 if height is counted by edges, and 0 if counted by nodes
    
    left_h = tree_height(root.left)
    right_h = tree_height(root.right)
    
    return 1 + max(left_h, right_h)

  # end



# Driver Code
if __name__ == '__main__':

	#Q1:
    root = None 

    # insert all nodes in the given order	by calling insert function
    nodes = [50,30,20,40,70,60,80,71,25,85]

    for value in nodes:
      root = insert(root, value)
    
    print("Inorder traversal of the BST:")
    inorder(root)   # print the tree sorted
    print() # newline





	#Q2:
    print("\n tree_height:")
    print(tree_height(root))

  #Q3:
    print("\n Number of leaf nodes:")
    print(count_leaf_nodes(root))

	#Q4:
    print("\n print nodes in level order:")
    print_nodes_bylevel(root)
    print()


  #Q5:
    print("\n preorder traversal of tree:")
    preorder(root)
    print()


  #Q6:
    print("\n print_nodes_with_two_child:")
    print_nodes_with_two_child(root)
    print()
```
